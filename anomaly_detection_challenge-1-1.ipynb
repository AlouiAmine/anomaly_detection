Anomaly detection (AD) refers to the process of detecting data points that do not conform with the rest of observations. Applications of anomaly detection include fraud and fault detection, surveillance, diagnosis, data cleanup, predictive maintenance.

When we talk about AD, we usually look at it as an unsupervised (or semi-supervised) task, where the concept of anomaly is often not well defined or, in the best case, just few samples are labeled as anomalous. In this challenge, you will look at AD from a different perspective!

The dataset you are going to work on consists of monitoring data generated by IT systems; such data is then processed by a monitoring system that executes some checks and detects a series of anomalies. This is a multi-label classification problem, where each check is a binary label corresponding to a specific type of anomaly. Your goal is to develop a machine learning model (or multiple ones) to accurately detect such anomalies.

This will also involve a mixture of data exploration, pre-processing, model selection, and performance evaluation. You will also be asked to try one or more rule learning models, and compare them with other ML models both in terms of predictive performances and interpretability. Interpreatibility is indeed a strong requirement especially in applications like AD where understanding the output of a model is as important as the output itself.

To use GMM, we first see how many components is best suited for each distribution. Therefore we check Akaike information criterion (AIC) which is an estimator of the best number of components to use.

```python
#Fill features with GMM
n_components = [10,20,30,40,50]
columns = ['CPU','PhysMEM','InstanceMEM','IndexServerAllocationLimit','Disk','LargestTableSize','LargestPartitionSize']
i = 1
plt.figure(figsize=(25,12))
for col in columns:
    aics = []
    print("Operating now: ",col)
    #For eachcomponents number we calculate AIC
    for n in n_components:
        model = GaussianMixture(n, covariance_type='full', random_state=0)
        tmp = train[train[col].notnull()]
        aics.append(model.fit(tmp[col].to_frame()).aic(tmp[col].to_frame()))
    plt.subplot(2,4,i)
    plt.plot(n_components, aics);
    plt.xlabel("Number of GMM components")
    plt.ylabel("AICS")
    plt.title(col)
    i=i+1;
```

```python
comp_size = [40,30,40,40,20,40,30]
columns = ['CPU','PhysMEM','InstanceMEM','IndexServerAllocationLimit','Disk','LargestTableSize','LargestPartitionSize']
#For each column we fill missing data with samples generated with the pre-computed distributions
for i in range(len(columns)):
    col = columns[i]
    tmp = train[train[col].notnull()]
    #Number of missing data in column
    num = len(train) - len(train[col].dropna())
    #Calculate distribution with GMM using the best number of components
    model = GaussianMixture(comp_size[i], covariance_type='full', random_state=0)
    model.fit(tmp[col].to_frame())
    #Sample and fill columns
    train.loc[train[col].isnull(), col] =model.sample(n_samples=num)[0]
```

We will first use the Bayesian Rule Set(RBS). Therefore we create two dataframes one containing all the features and the other containing all the outputs.

```python


columns = ['Check1','Check2', 'Check3', 'Check4', 'Check5','Check6', 'Check7', 'Check8']
for col in columns:
    print("________Calculation for ", col, " started________")
    t = time()
    n = 10000
    
    y_solo=y_train[col][:n].as_matrix()
    model = BayesianRuleSet(method='forest',max_rules=10,max_iter=10)
    model.fit(X_train[:n], y_solo.astype(int))
    yhat = model.predict(X_test[:n])
    print(classification_report(y_test[col].astype(int)[:n],yhat))
    print("Time of execution: ", str(time()-t))
    
    
```

```python
import xgboost
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from xgboost import plot_importance
ls_model=[]
for i in range(8):
    print("________Calculation for Check" +str(i+1) +" started________")
    start=time()
    X_train_, X_val_, y_train_, y_val_ = train_test_split(X_train,y_train['Check'+str(i+1)], test_size=0.2, random_state=42)

    xgtrain = xgboost.DMatrix(X_train_, y_train_)
    xgtest = xgboost.DMatrix(X_val_, y_val_)
    eta = 0.1
    max_depth = 1
    subsample = 0.8
    colsample_bytree = 0.8
    params = {
        "objective": "binary:logistic",
        "booster" : "gbtree",
        "eval_metric": "logloss",
        "eta": eta,
        "max_depth": max_depth,
        "subsample": subsample,
        "colsample_bytree": colsample_bytree,
        "silent": 1,
        "seed": 42
    }
    model = xgboost.train(
        params=list(params.items()),  
        early_stopping_rounds=80, 
        verbose_eval=50, 
        dtrain=xgtrain,
        evals=[(xgtrain, 'train'), (xgtest, 'test')],
        num_boost_round=100,
    )

    X_test_=xgboost.DMatrix(X_val_)
    p=model.predict(X_test_)
    p=[1 if p[i]>0.5 else 0 for i in range(len(p))]



    print('Confusion Matrix:\n')
    print(confusion_matrix(y_val_, p))
    print('\n classification Report:\n')
    report = classification_report(y_pred=y_val_, y_true=p)
    print(report)
    plot_importance(model)
    plt.show()
    print('\n time execution : {} \n'.format(time()-start))
    ls_model.append(model)
```
